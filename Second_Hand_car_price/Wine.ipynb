```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import LinearSVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import scipy.stats as st
from sklearn.model_selection import cross_val_score
#sklearn.metrics.SCORERS.keys()
```

```python
df= pd.read_csv('wines_SPA.csv')
df
```

```python
df.isna().sum()
df.dropna(inplace=True)
df.isna().sum()
```

```python

df= df[df['year']!='N.V.']
```

```python
df=pd.get_dummies(df,columns=['type'])
df
```

```python
df['acidity'].unique()
```

```python
df['year']=df['year'].astype(int)
#df['Review_Quality'] = df['rating']*df['num_reviews']
```

```python
ds = df['num_reviews']
#sn.scatterplot(x="SIZE",y="PRICE",data=ds)
Q1=ds.quantile(q=0.25)#Remove OUTLIERS through IQR for better model
Q2=ds.quantile(q=.75)
IQR=ds.apply(st.iqr)
X=ds[~((ds<(Q1-1.5*IQR))|((ds>Q2+1.5*IQR)))]#Interquartile calculation
```

```python
df['new']=df.index.isin(X.index)
df=df[df['new']==True]
df.isna().sum().sum()
```

```python
#X= df[['year','rating','num_reviews','body','acidity','type_Rioja Red','type_Rioja Red']]
X=df.drop(['winery','wine','country','region','price','new'],axis=1)
#X=np.array(df[['year','type_Rioja White']])
y=df['price']


Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2)
list(Xtrain.columns)
```

```python
score=[]

def regg(Model):
    model = Model
    model.fit(Xtrain,ytrain)
    k= cross_val_score(model,X=X,y=y,scoring='r2').mean()
    score.append(k)
    

    
models= [LinearRegression,RandomForestRegressor,LinearSVR,KNeighborsRegressor]
for i in models:
    regg(i())
score
```

```python
Regg_model= RandomForestRegressor()
Regg_model.fit(Xtrain,ytrain)
```
